```python
import pandas as pd
import numpy as np
```

```python
filename = 'california_raw_utf8.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
amps_lst = []
kw_lst = []
```

```python
def unit_value(speed, unit):
  speed = speed.replace(unit, '')
  speed = speed.replace(' ', '')

  if '?' in speed:
    return 'NULL'
  else:
    return speed
```

```python
for i in range(len(df)):
  try:
    charge_speed = df['CHARGE_SPEED'][i]
    amps, kw = charge_speed.split(',')

    amps = unit_value(amps, 'A')
    amps_lst.append(amps)

    kw = unit_value(kw, 'kW')
    kw_lst.append(kw)
    
  except:
    amps_lst.append('?')
    kw_lst.append('?')
```

```python
print(len(amps_lst))
```

```python
df['AMPS'] = amps_lst
df['KW'] = kw_lst
```

```python
df.head()
```

```python
df.to_csv(f'{filename}_v1.csv', index=False)
```

```python
price_dict = {}
```

```python
for i in range(len(df)):
  price_name = df['PRICE_RAW'][i]

  if price_name in price_dict:
    price_dict[price_name] += 1
  else:
    price_dict[price_name] = 1
```
