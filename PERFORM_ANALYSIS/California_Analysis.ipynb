```python
import pandas as pd
import numpy as np
```

LOAD FILE

```python
filename = 'california_7-5-22.csv'

df = pd.read_csv(filename)
df['CITY'] = df['CITY'].str.title()
```

CALIFORNIA ANALYSIS

```python
df = df[(df['NON_KWH'] > 0) | (df['MEM_KWH'] > 0)]
df_california = df.reset_index()

# ultra-fast charging >= 125 kw
df_california_ufc = df_california[df_california['KW'] >= 125]
df_california_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_california_stage = df_california[df_california['KW'] >= 50]
df_california_fc = df_california_stage[df_california_stage['KW'] < 125]
df_california_fc.reset_index(drop=True, inplace=True)

print('california fc non:', df_california_fc['NON_KWH'].mean())
print('california fc mem:', df_california_fc['MEM_KWH'].mean())
print('california ufc non:', df_california_ufc['NON_KWH'].mean())
print('california ufc mem:', df_california_ufc['MEM_KWH'].mean())
```

CALIFORNIA BY NETWORK

```python
# Get network name
fc_charging_network_california = []
for i in range(len(df_california_fc)):
  network_name = df_california_fc['NETWORK'][i]
  if network_name not in fc_charging_network_california:
    fc_charging_network_california.append(network_name)

ufc_charging_network_california = []
for i in range(len(df_california_ufc)):
  network_name = df_california_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_california:
    ufc_charging_network_california.append(network_name)

print('california fc:', fc_charging_network_california)
print('california ufc:', ufc_charging_network_california)
```
