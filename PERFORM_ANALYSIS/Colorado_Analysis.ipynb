```python
import pandas as pd
import numpy as np
```

LOAD FILE

```python
filename = 'colorado_7-1-22.csv'

df = pd.read_csv(filename)
df['CITY'] = df['CITY'].str.title()
```

COLORADO ANALYSIS

```python
df = df[(df['NON_KWH'] > 0) | (df['MEM_KWH'] > 0)]
df_colorado = df.reset_index()

# ultra-fast charging >= 125 kw
df_colorado_ufc = df_colorado[df_colorado['KW'] >= 125]
df_colorado_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_colorado_stage = df_colorado[df_colorado['KW'] >= 50]
df_colorado_fc = df_colorado_stage[df_colorado_stage['KW'] < 125]
df_colorado_fc.reset_index(drop=True, inplace=True)

print('colorado fc non:', df_colorado_fc['NON_KWH'].mean())
print('colorado fc mem:', df_colorado_fc['MEM_KWH'].mean())
print('colorado ufc non:', df_colorado_ufc['NON_KWH'].mean())
print('colorado ufc mem:', df_colorado_ufc['MEM_KWH'].mean())
```

COLORADO BY NETWORK

```python
# Get network name
fc_charging_network_colorado = []
for i in range(len(df_colorado_fc)):
  network_name = df_colorado_fc['NETWORK'][i]
  if network_name not in fc_charging_network_colorado:
    fc_charging_network_colorado.append(network_name)

ufc_charging_network_colorado = []
for i in range(len(df_colorado_ufc)):
  network_name = df_colorado_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_colorado:
    ufc_charging_network_colorado.append(network_name)

print('colorado fc:', fc_charging_network_colorado)
print('colorado ufc:', ufc_charging_network_colorado)
```

```python
df_colorado_fc_evgo = df_colorado_fc[df_colorado_fc['NETWORK'] == 'EVgo']
df_colorado_fc_cp = df_colorado_fc[df_colorado_fc['NETWORK'] == 'ChargePoint']
df_colorado_fc_evc = df_colorado_fc[df_colorado_fc['NETWORK'] == 'EVConnect']
df_colorado_fc_unknown = df_colorado_fc[df_colorado_fc['NETWORK'] == 'Unknown']
df_colorado_ufc_ea = df_colorado_ufc[df_colorado_ufc['NETWORK'] == 'Electrify America']

print('colorado evgo non:', df_colorado_fc_evgo['NON_KWH'].mean())
print('colorado evgo mem:', df_colorado_fc_evgo['MEM_KWH'].mean())

print('colorado cp non:', df_colorado_fc_cp['NON_KWH'].mean())
print('colorado cp mem:', df_colorado_fc_cp['MEM_KWH'].mean())

print('colorado evc non:', df_colorado_fc_evc['NON_KWH'].mean())
print('colorado evc mem:', df_colorado_fc_evc['MEM_KWH'].mean())

print('colorado unknown non:', df_colorado_fc_unknown['NON_KWH'].mean())
print('colorado unknown mem:', df_colorado_fc_unknown['MEM_KWH'].mean())

print('colorado ea non:', df_colorado_ufc_ea['NON_KWH'].mean())
print('colorado ea mem:', df_colorado_ufc_ea['MEM_KWH'].mean())
```

DENVER ANALYSIS

```python
df_denver = df[(df['CITY'] == 'Aurora') | (df['CITY'] == 'Commerce City') | (df['CITY'] == 'Englewood') |
                (df['CITY'] == 'Littleton') | (df['CITY'] == 'Watkins') | (df['CITY'] == 'Denver')]

df_denver.reset_index(drop=True, inplace=True)

# ultra-fast charging >= 125 kw
df_denver_ufc = df_denver[df_denver['KW'] >= 125]
df_denver_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_denver_stage = df_denver[df_denver['KW'] >= 50]
df_denver_fc = df_denver_stage[df_denver_stage['KW'] < 125]
df_denver_fc.reset_index(drop=True, inplace=True)

print('denver fc non:', df_denver_fc['NON_KWH'].mean())
print('denver fc mem:', df_denver_fc['MEM_KWH'].mean())
print('denver ufc non:', df_denver_ufc['NON_KWH'].mean())
print('denver ufc mem:', df_denver_ufc['MEM_KWH'].mean())
```

DENVER BY NETWORK

```python
# Get network name
fc_charging_network_denver = []
for i in range(len(df_denver_fc)):
  network_name = df_denver_fc['NETWORK'][i]
  if network_name not in fc_charging_network_denver:
    fc_charging_network_denver.append(network_name)

ufc_charging_network_denver = []
for i in range(len(df_denver_ufc)):
  network_name = df_denver_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_denver:
    ufc_charging_network_denver.append(network_name)

print('denver fc:', fc_charging_network_denver)
print('denver ufc:', ufc_charging_network_denver)
```

```python
df_denver_fc_evgo = df_denver_fc[df_denver_fc['NETWORK'] == 'EVgo']
df_denver_fc_cp = df_denver_fc[df_denver_fc['NETWORK'] == 'ChargePoint']
df_denver_fc_evc = df_denver_fc[df_denver_fc['NETWORK'] == 'EVConnect']
df_denver_fc_unknown = df_denver_fc[df_denver_fc['NETWORK'] == 'Unknown']
df_denver_ufc_ea = df_denver_ufc[df_denver_ufc['NETWORK'] == 'Electrify America']

print('denver evgo non:', df_denver_fc_evgo['NON_KWH'].mean())
print('denver evgo mem:', df_denver_fc_evgo['MEM_KWH'].mean())

print('denver cp non:', df_denver_fc_cp['NON_KWH'].mean())
print('denver cp mem:', df_denver_fc_cp['MEM_KWH'].mean())

print('denver evc non:', df_denver_fc_evc['NON_KWH'].mean())
print('denver evc mem:', df_denver_fc_evc['MEM_KWH'].mean())

print('denver unknown non:', df_denver_fc_unknown['NON_KWH'].mean())
print('denver unknown mem:', df_denver_fc_unknown['MEM_KWH'].mean())

print('denver ea non:', df_denver_ufc_ea['NON_KWH'].mean())
print('denver ea mem:', df_denver_ufc_ea['MEM_KWH'].mean())
```
