```python
import pandas as pd
import numpy as np
```

LOAD FILE

```python
filename = 'texas_7-1-22.csv'

df = pd.read_csv(filename)
df['CITY'] = df['CITY'].str.title()
```

TEXAS ANALYSIS

```python
df = df[(df['NON_PER_MIN'] > 0) | (df['MEM_PER_MIN'] > 0)]
df_texas = df.reset_index()

# ultra-fast charging >= 125 kw
df_texas_ufc = df_texas[df_texas['KW'] >= 125]
df_texas_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_texas_stage = df_texas[df_texas['KW'] >= 50]
df_texas_fc = df_texas_stage[df_texas_stage['KW'] < 125]
df_texas_fc.reset_index(drop=True, inplace=True)

print('texas fc non:', df_texas_fc['NON_PER_MIN'].mean())
print('texas fc mem:', df_texas_fc['MEM_PER_MIN'].mean())
print('texas ufc non:', df_texas_ufc['NON_PER_MIN'].mean())
print('texas ufc mem:', df_texas_ufc['MEM_PER_MIN'].mean())
```

TEXAS BY NETWORK

```python
# Get network name
fc_charging_network_texas = []
for i in range(len(df_texas_fc)):
  network_name = df_texas_fc['NETWORK'][i]
  if network_name not in fc_charging_network_texas:
    fc_charging_network_texas.append(network_name)

ufc_charging_network_texas = []
for i in range(len(df_texas_ufc)):
  network_name = df_texas_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_texas:
    ufc_charging_network_texas.append(network_name)

print('texas fc:', fc_charging_network_texas)
print('texas ufc:', ufc_charging_network_texas)
```

```python
df_texas_fc_evgo = df_texas_fc[df_texas_fc['NETWORK'] == 'EVgo']
df_texas_fc_cp = df_texas_fc[df_texas_fc['NETWORK'] == 'ChargePoint']
df_texas_fc_evc = df_texas_fc[df_texas_fc['NETWORK'] == 'EVConnect']
df_texas_ufc_ea = df_texas_ufc[df_texas_ufc['NETWORK'] == 'Electrify America']

print('texas evgo non:', df_texas_fc_evgo['NON_PER_MIN'].mean())
print('texas evgo mem:', df_texas_fc_evgo['MEM_PER_MIN'].mean())

print('texas cp non:', df_texas_fc_cp['NON_PER_MIN'].mean())
print('texas cp mem:', df_texas_fc_cp['MEM_PER_MIN'].mean())

print('texas evc non:', df_texas_fc_evc['NON_PER_MIN'].mean())
print('texas evc mem:', df_texas_fc_evc['MEM_PER_MIN'].mean())

print('texas ea non:', df_texas_ufc_ea['NON_PER_MIN'].mean())
print('texas ea mem:', df_texas_ufc_ea['MEM_PER_MIN'].mean())
```
