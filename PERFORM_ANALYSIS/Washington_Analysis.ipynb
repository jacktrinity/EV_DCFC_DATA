```python
import pandas as pd
import numpy as np
```

LOAD FILE

```python
filename = 'washington_7-6-22.csv'

df = pd.read_csv(filename)
df['CITY'] = df['CITY'].str.title()
```

WASHINGTON ANALYSIS

```python
df = df[(df['NON_KWH'] > 0) | (df['MEM_KWH'] > 0)]
df_washington = df.reset_index()

# ultra-fast charging >= 125 kw
df_washington_ufc = df_washington[df_washington['KW'] >= 125]
df_washington_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_washington_stage = df_washington[df_washington['KW'] >= 50]
df_washington_fc = df_washington_stage[df_washington_stage['KW'] < 125]
df_washington_fc.reset_index(drop=True, inplace=True)

print('washington fc non:', df_washington_fc['NON_KWH'].mean())
print('washington fc mem:', df_washington_fc['MEM_KWH'].mean())
print('washington ufc non:', df_washington_ufc['NON_KWH'].mean())
print('washington ufc mem:', df_washington_ufc['MEM_KWH'].mean())
```

WASHINGTON BY NETWORK

```python
# Get network name
fc_charging_network_washington = []
for i in range(len(df_washington_fc)):
  network_name = df_washington_fc['NETWORK'][i]
  if network_name not in fc_charging_network_washington:
    fc_charging_network_washington.append(network_name)

ufc_charging_network_washington = []
for i in range(len(df_washington_ufc)):
  network_name = df_washington_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_washington:
    ufc_charging_network_washington.append(network_name)

print('washington fc:', fc_charging_network_washington)
print('washington ufc:', ufc_charging_network_washington)
```

```python
df_washington_fc_evgo = df_washington_fc[df_washington_fc['NETWORK'] == 'EVgo']
df_washington_fc_cp = df_washington_fc[df_washington_fc['NETWORK'] == 'ChargePoint']
df_washington_fc_sr = df_washington_fc[df_washington_fc['NETWORK'] == 'Shell Recharge']
df_washington_fc_evc = df_washington_fc[df_washington_fc['NETWORK'] == 'EVConnect']
df_washington_ufc_ea = df_washington_ufc[df_washington_ufc['NETWORK'] == 'Electrify America']
df_washington_ufc_evc = df_washington_ufc[df_washington_ufc['NETWORK'] == 'EVConnect']


print('washington evgo non:', df_washington_fc_evgo['NON_KWH'].mean())
print('washington evgo mem:', df_washington_fc_evgo['MEM_KWH'].mean())

print('washington cp non:', df_washington_fc_cp['NON_KWH'].mean())
print('washington cp mem:', df_washington_fc_cp['MEM_KWH'].mean())

print('washington sr non:', df_washington_fc_sr['NON_KWH'].mean())
print('washington sr mem:', df_washington_fc_sr['MEM_KWH'].mean())

print('washington evc non:', df_washington_fc_evc['NON_KWH'].mean())
print('washington evc mem:', df_washington_fc_evc['MEM_KWH'].mean())

print('washington ea non:', df_washington_ufc_ea['NON_KWH'].mean())
print('washington ea mem:', df_washington_ufc_ea['MEM_KWH'].mean())

print('washington evc non:', df_washington_ufc_evc['NON_KWH'].mean())
print('washington evc mem:', df_washington_ufc_evc['MEM_KWH'].mean())
```

SEATTLE ANALYSIS

```python
df_seattle = df[(df['CITY'] == 'Kirkland') | (df['CITY'] == 'Seattle')]

df_seattle.reset_index(drop=True, inplace=True)

# ultra-fast charging >= 125 kw
df_seattle_ufc = df_seattle[df_seattle['KW'] >= 125]
df_seattle_ufc.reset_index(drop=True, inplace=True)

# fast charging >= 50 kw
df_seattle_stage = df_seattle[df_seattle['KW'] >= 50]
df_seattle_fc = df_seattle_stage[df_seattle_stage['KW'] < 125]
df_seattle_fc.reset_index(drop=True, inplace=True)

print('seattle fc non:', df_seattle_fc['NON_KWH'].mean())
print('seattle fc mem:', df_seattle_fc['MEM_KWH'].mean())
print('seattle ufc non:', df_seattle_ufc['NON_KWH'].mean())
print('seattle ufc mem:', df_seattle_ufc['MEM_KWH'].mean())
```

SEATTLE BY NETWORK

```python
# Get network name
fc_charging_network_seattle = []
for i in range(len(df_seattle_fc)):
  network_name = df_seattle_fc['NETWORK'][i]
  if network_name not in fc_charging_network_seattle:
    fc_charging_network_seattle.append(network_name)

ufc_charging_network_seattle = []
for i in range(len(df_seattle_ufc)):
  network_name = df_seattle_ufc['NETWORK'][i]
  if network_name not in ufc_charging_network_seattle:
    ufc_charging_network_seattle.append(network_name)

print('seattle fc:', fc_charging_network_seattle)
print('seattle ufc:', ufc_charging_network_seattle)
```
